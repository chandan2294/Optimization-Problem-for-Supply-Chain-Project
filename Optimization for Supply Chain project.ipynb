# **Optimzation model to find number of workers required to fulfill the forecasted sales**

After we forecast the number of sales for next 6 months, we need to find out the number of workers required by the company to fulfil those requirements so that the compant doesn't overhire or underhire. We can build an optimization model to acheive this task:

I- weekday 
* 1-Monday 
* 2-Tuesday 
* 3 - Wednesday
* 4 - Thursday
* 5 - Friday
* 6 - Saturday
* 7 - Sunday

I: {1,2,….7}

J- Shift of the day : 
* 1- 12am to 8 am 
* 2- 8am to 4 pm
* 3 - 4pm to 12am		

J: {1,2,3}

## Parameters 

$A_{ij}$- Forecasted total number of orders per shift *j* on day *i*	
i ∈ I, j ∈ J	

$D_{ij}$- Forecasted distance of the each order of shift *j* on day *i*

## Variables

$Y_{ij}$ – Number of workers who will work on day *I* on shift *j*

$C_{ij}$ - Total hours of overtime require on day *i* on shift *j*

$W_{ij}$ - Number of workers who will do overtime on day *I* on shift *j*

**Objective Function**

**Min Z** = $\sum_{i=1}^7$ $\sum_{j=1}^3$ [((25 + (12.5 $\times$ $D_{ij}$ )) $\times$ $A_{ij} \times$ 0.2) - (100 $\times  C_{ij}$)]

*Subject To,*

Workers need 30 min break in 8 hrs shift 

$A_{ij}$(13+(4$\times D_{ij}$)) = (450$\times Y_{ij}$) + ($C_{ij} \times$ 60) 
*For every i ∈ I, j  ∈ J*

Total Number of workers who can work on every day should not exceed 227

$\sum_{j=1}^3 Y_{ij} \leq 227$	*For every i ∈ I*

Overtime workers can’t work more than 3 hrs

3$\times W_{ij} \geq C_{ij}$		*For every i ∈ I, j  ∈ J*

Number of overtime workers should be less than 20 on every shift

$W_{ij} \geq 20$		*For every i ∈ I, j  ∈ J*


## **Objective Function**

**Min Z= $\sum_{i=1}^7 \sum_{j=1}^3 X_{ij}$**

Subject To,

$A_{ij}$(13+(4$\times D_{ij}$)) $\leq$ (450$\times X_{ij}$)	

$Y_i$ = $\sum_{j=1}^3 X_{ij}$

## **Optimization Part Starts here**

**Optimization models:**

The number of workers required per shift on each day of week for the forecasted number of orders & average distance per order is found by using the optimization model. And, further to find the number of regular workers & overtime workers on each shift is also found by using the another optimization model. First optimization model will help industry to work with Level strategy. Level strategy suggest industry to hire constant workforce to complete orders. It will give average number of workers required per day, as company have to maintain constant total workforce per day. Above two optimization models are as below. 

* Optimization model to find total workforce required on each shift of each day of week

Optimization model will give the number of workers required per shift on each day of week for the forecasted average number of orders and average distance per. So, basically this model will use results of forecasting. This model is as below,

*Variables*, 

$X_{ij}$ – number of workers required on day i on shift j          ……for every i ∈ I, j  ∈ J

$Y_i$ – total number of workers required on day i                   ……for every i ∈ I, j  ∈ J

*Parameters*,

$A_{ij}$ - forecasted total number of orders per shift j on day i        ……for every i ∈ I, j  ∈ J		

$D_{ij}$- forecasted distance of the each order of shift j on day i    ……for every i ∈ I, j  ∈ J

*Objective Function*,

**Min        Z = $\sum_{i=1}^7 \sum_{j=1}^3 X_{ij}$**

*Subject to*,
	        		
1. $A_{ij}$(13+(4$\times D_{ij}$)) =< (450$\times X_{ij}$)                            ……for every i ∈ I, j  ∈ J
2. $Y_i$ = $\sum_{j=1}^3 X_{ij}$                             ……for every i ∈ I

3. Yi ∈ Z+,     Xij ∈ Z+

I- weekday: 1-Monday, 2-Tuesday...... 					I: {1, 2,….7}

J- Shift of the day: 1- 12am to 8 am, 2- 8am to 4 pm, 3 - 4pm to 12am    J: {1,2,3}

In this model, we are minimizing the total number of workers required per on each shift of each day of week. So objective function is minimizing the workforce. The constraint (1) represents that every worker requires the 30 minutes break in every shift of 8 hrs, as there’s rule such as company must give 30 minutes break if worker is working more than 6 hrs. So this rule limits worker’s availability on every shift, i.e. worker is available only for (8hrs – 30 min) = 450 minutes on every shift. Another constrain (2) sums the total number of workers required per day. And variables $Y_i$ & $X_ij$ both are positive integers.

* Optimization model to find regular & overtime workforce required on each shift of each day of week

From above model we got the average number of total worker’s company should hire i.e. 229 for a day. So for the days where company require more than 229 workers company will ask regular workers to do overtime. So for finding the number of regular & overtime workers on each shift on each day of week, the optimization model is designed which is as follows,

*Variables*,

$Y_{ij}$ - Number of workers who will work on day I on shift j		……for every i ∈ I, j  ∈ J

$C_{ij}$  - Total hours of overtime require on day i on shift j	            ……for every i ∈ I, j  ∈ J

$W_{ij}$  - Number of workers who will do overtime on day I on shift j   ……for every i ∈ I, j  ∈ J

*Parameters*,

$A_{ij}$ - forecasted total number of orders per shift j on day i        ……for every i ∈ I, j  ∈ J		

$D_{ij}$ - forecasted distance of the each order of shift j on day i    ……for every i ∈ I, j  ∈ J

*Objective Function,*

**Max Z = $\sum_{i=1}^7 \sum_{j=1}^3$ [((25 + (12.5 $\times D_{ij}$)) $\times A_{ij} \times$ 0.2) - (100 - $C_{ij}$]**

*Subject to,*

1. $A_{ij}$(13+(4$\times D_{ij}$)) = (450$\times Y_{ij}$)+($C_{ij} \times$60)		For every i ∈ I, j  ∈ J
2. $\sum_{j=1}^3 Y_{ij} \leq$ 229				For every i ∈ I
3. 3$\times W_{ij} \geq C_{ij}$ 				For every i ∈ I, j  ∈ J
4. $W_{ij} \leq$ 20				          For every i ∈ I, j  ∈ J

Yij ∈ Z+,     Wij ∈ Z+,   Cij>=0

I- weekday: 1-Monday, 2-Tuesday...... 					I: {1, 2,….7}

J- Shift of the day: 1- 12am to 8 am, 2- 8am to 4 pm, 3 - 4pm to 12am    J: {1,2,3}


The objective function of the above optimization model is to maximize the profit of the industry. The objective function is the subtraction of the 20% of the amount paid by customer for every order and the number of overtime hours required on every shift multiplied by 100, because company pays 100 bucks per hour to worker if he is doing overtime. Next, is the constraint (1) which represents the workers are available only for 450 minutes in 8 hrs shift. Constraint (2) is for only 229 workers can work together & constraint (3) says worker can’t do overtime more than 3 hrs. The total overtime workers per shift should not exceed 20, which is stated by constraint (4).

```python
from gurobipy import *
import math as mt
I=[1,2,3,4,5,6,7]
J=[1,2,3]
A={1:{1:624.08,2:1248.16,3:1248.16},2:{1:710.04,2:1420.08,3:1420.08},3:{1:753.36,2:1506.72,3:1506.72},4:{1:774.04,2:1548.08,3:1548.08},5:{1:815.28,2:1426.74,3:1834.38},6:{1:1033.15,2:1653.04,3:1446.41},7:{1:1224.9,2:1224.9,3:1633.2}}
D={1:{1:3.51775,2:3.51775,3:3.51775},2:{1:3.36302,2:3.36302,3:3.36302},3:{1:3.53440,2:3.53440,3:3.53440},4:{1:3.7,2:3.7,3:3.7},5:{1:3.44347,2:3.44347,3:3.44347},6:{1:3.44978,2:3.44978,3:3.44978},7:{1:3.32686,2:3.32686,3:3.32686}}
# I- weekday : 1-Monday, 2-Tuesday......
#J- Shift of the day : 1- 12am to 8 am , 2- 8am to 4 pm, 3 - 4pm to 12am
# A- forecasted total number of orders per shift j on day i
# D- forecasted distance of the each order of shift j on day i
```

```python
def min_time(A,D,I,J):
    m = Model("Problem") 
    # x[i,j] : number of workers required on shift j of day i without considering overtime
    x={}
    for i in I:
        for j in J:
            x[i,j]=m.addVar(vtype=GRB.CONTINUOUS,name='x_%d%d' %(i,j))

    # y[i] : total number of workers required per day i
    y={}
    for i in I:
        for j in J:
            y[i,j]=m.addVar(vtype=GRB.INTEGER,name='y_%d%d' %(i,j))

    # c[i,j] : total hrs rquired per shift j on day i for overtime
    c={}
    for i in I:
        for j in J:
            c[i,j]=m.addVar(vtype=GRB.CONTINUOUS,name='c_%d%d' %(i,j))

    # w[i,j] : total number of workers required per shift j on day i which will do overtime
    w={}
    for i in I:
        for j in J:
            w[i,j]=m.addVar(vtype=GRB.INTEGER,name='w_%d%d' %(i,j))

    # Objective value
    z={}
    z=m.addVar(vtype=GRB.CONTINUOUS, name='z', obj=1)

    h={}
    h=m.addVar(vtype=GRB.CONTINUOUS, name='z', obj =1)

    m.update()

    # Objective Function

    m.addConstr(z==quicksum(((((12.5*D[i][j])+25)*A[i][j])*0.2-(100*w[i,j]*3))for i in I for j in J))
    m.setObjective(z, GRB.MAXIMIZE)
    m.addConstr(h==quicksum((w[i,j])for i in I for j in J))
    m.setObjective(h, GRB.MINIMIZE)

    # Constraints
    # Constraint 1: Every worker can work only for 8 hrs in each shift with 30 min break & for every order time required
    # time required per order is sum of constant time + time required per km
    # total time required for completing orders on day i for shioft j is sum of regular workers & overtime workers with their allowed time limit
    for i in I:
        for j in J:
            m.addConstr((A[i][j]*(13+(4*D[i][j])))==((450*y[i,j])+(c[i,j]*60)))

    #To find total workers required without overtime
    for i in I:
        for j in J:
            m.addConstr((A[i][j]*(13+(4*D[i][j])))==(450*x[i,j]))

    # Constraint 2: total number of workers working in day should less than equal to total number of employees hired  i.e. 227
    for i in I:
        m.addConstr(227>=quicksum((y[i,j])for j in J))

    # Constraint 3: If worker is doing overtime then he is allowed to do overtime only upto 3hrs
    for i in I:
        for j in J:
            m.addConstr(3*w[i,j]>=c[i,j])

    # Constraint 4: Total number of overtime workers per shift should no exceed 20 count
    for i in I:
        for j in J:
            m.addConstr(w[i,j]<=20)
```

```python
def printSolution():
    if m.status==GRB.status.OPTIMAL:
        print("Min time for project ", m.objVal)
        for i in I:
            for j in J:
                print(x[i,j])
        for i in I:
            for j in J:
                print(y[i,j])
        for i in I:
            for j in J:
                print(c[i,j])
        for i in I:
            for j in J:
                print(w[i,j])
    print(z)
```

```python
min_time(A,D,I,J)
m.optimize()
printSolution()
```
